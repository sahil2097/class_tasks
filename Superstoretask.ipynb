```python
#9 . base on question number 8 find out for which order id we have shipment duration more than 10 days 

def delayed_orders(a):
    if a > 10:
        return 'delayed'
    else:
        return 'on_time'
```

```python
df['shipping_time'] =df['shipping_time'].astype(str)
df['shipping_time'] = df['shipping_time'].str.replace("days","")
df['shipping_time'] = df['shipping_time'].astype(int)
df['processing_time'] = df['shipping_time'].apply(delayed_orders)
```

```python
df[df['shipping_time']>10][['Order ID','processing_time']]
```
